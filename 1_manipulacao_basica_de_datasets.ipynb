##### Criando um DataFrame

A forma mais comum é importar um DataFrame, porém também é possível criar um a partir de um dicionário.<br>

As **chaves** representam as **colunas** e **cada um dos valores de sua lista** representa o **valor da linha** correspondente para aquela coluna.

```python
star_wars_dict = {
    "name": ["Luke Skywalker", "Yoda", "Palpatine"],
    "age": [16, 1000, 70],
    "weight_kg": [70, 15, 60],
    "is_jedi": [True, True, False]
}

star_wars = pd.DataFrame(star_wars_dict)
star_wars
```

##### Index-based selection (seleção baseada em Índices)

Mostrando linhas específicas de um DataFrame:

`iloc`: seleciona elementos do Dataframe, baseado em seu **índice (número)** --> row-first, column-second

###### **Selecionando uma linha/observação:**

```python
# Selecionando a linha 1 - observação de index 1 do DataFrame
data.iloc[1]
```

###### **Selecionando múltiplas linhas/observações:**

```python
# Selecionando as linhas de index de 0 a 5 (slice, continuo)
data.iloc[:6]
```

```python
# Selecionar as linhas de index 1, 5, 10 e 15 (isoladamente, ordem não importa)
data.iloc[[1, 5, 10, 15]]
```

```python
# Selecionar um valor baseado no index e coluna
# index 1, coluna 2 ("REGIÃO")
data.iloc[1, 2] 
```

##### Label-based selection (seleção baseada em rótulos)
