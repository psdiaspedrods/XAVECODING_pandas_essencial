# Estatisticas Destritivas

```python
import pandas as pd
```

```python
data = pd.read_csv('./datasets/gas_prices_brazil_2004-2021_pre.csv')
```

```python
data
```

O Pandas fornecem algumas funções/métodos que computam certas estatísticas descritivas.

`describe`: exibe várias **estatísticas descritivas** para os _atributos_ de um dataframe ou para uma _Series_.

```python
stats = data.describe()
stats
```

```python
stats[["PREÇO MÉDIO REVENDA", "PREÇO MÍNIMO REVENDA", "MARGEM MÉDIA REVENDA"]]
```

```python
stats.loc[["min", "max", "mean"]]
```

```python
stats.loc[["min", "max", "mean"], ["PREÇO MÉDIO REVENDA", "PREÇO MÍNIMO REVENDA"]]
```

Existe a possibilidade de calcular individualmente

```python
data["PREÇO MÍNIMO REVENDA"].min()
```

```python
data["PREÇO MÍNIMO REVENDA"].mean()
```

```python
data["PREÇO MÍNIMO REVENDA"].max()
```

```python
data["PREÇO MÍNIMO REVENDA"].std()
```

`.value_counts()`:  Conta a frequência dos valores de uma dada variável (de preferência, _categórica_).

```python
data["ESTADO"].value_counts()
```

```python
data["ESTADO"].value_counts().to_frame()
```
